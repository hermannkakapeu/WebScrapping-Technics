# Scrapping Data from real web site + Pandas

```python
url1 = 'https://en.wikipedia.org/wiki/Christianity_by_country'

page = requests.get(url1)
soup1 = BeautifulSoup(page.text, 'html')
```

```python
print(soup1)
```

```python
table = soup1.find_all('table')[11]
head = table.find_all('th')
list_head = [title.text.strip() for title in head[:11] ]

list_head
```

```python
import pandas as pd

df = pd.DataFrame(columns=list_head)
df
```

```python
for row in table.find_all('tr')[1:]:
    dd = []
    for data in row.find_all('td') :
        dd.append(data.text.strip())
    print(dd)
    length = len(df)
    df.loc[length] = dd
    
```
